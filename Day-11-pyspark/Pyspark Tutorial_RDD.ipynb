```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
spark
```

```python
sc = spark.sparkContext
```

```python
sc
```

```python
nums = list(range(0, 1000001))
len(nums)
```

```python
nums_rdd = sc.parallelize(nums)
nums_rdd
```

```python
nums_rdd.take(5)
```

```python
squared_nums_rdd = nums_rdd.map(lambda x: x ** 2 )
```

```python
squared_nums_rdd.take(5)
```

```python
pairs = squared_nums_rdd.map(lambda x: (x, len(str(x))))
pairs.take(20)
```

```python
# filter transformation Return a new RDD containing only the elements that satisfy a predicate.
even_digit_pairs = pairs.filter(lambda x: x[1] % 2 == 0 )
even_digit_pairs.take(20)
```

```python
flipped_pairs = even_digit_pairs.map(lambda x: (x[1], x[0]))
flipped_pairs.take(20)
```

```python
grouped_flipped_pairs = flipped_pairs.groupByKey()
grouped_flipped_pairs.take(20)
```

```python
grouped = grouped_flipped_pairs.map(lambda x: (x[0], list(x[1])))
grouped.take(2)
```

```python
averaged = grouped.map(lambda x: (x[0], sum(x[1]) / len(x[1]) ))
averaged.collect()
```
